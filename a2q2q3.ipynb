```python
from geopy.geocoders import Nominatim

# Specify a custom user agent
geolocator = Nominatim(user_agent='bmi6016_dwa')

# Use the geolocator to get the location information for the address
address = '9301 Wood Rd, Riverside, CA 92508'
location = geolocator.geocode(address)

if location is not None:
    print((location.latitude, location.longitude))
else:
    print('Unable to geocode the address')
```

```python
from geopy import distance

# define the coordinates of the school
school_coords = (33.883578650000004, -117.33465430059044)

# calculate the distance between each monitor and the school
geo_df['distance'] = geo_df.apply(lambda row: distance.distance(school_coords, (row['SITE_LATITUDE'], row['SITE_LONGITUDE'])).km, axis=1)

# select the monitor with the smallest distance
nearest_monitor = geo_df.loc[geo_df['distance'].idxmin()]

# print the information about the nearest monitor
print("Nearest monitor information:")
print("---------------------------")
print("Site ID:", nearest_monitor['Site ID'])
print("Site Name:", nearest_monitor['Site Name'])
print("Distance (km):", nearest_monitor['distance'])
```

```python
# Filter the DataFrame to only include rows with the nearest monitor's Site ID
nearest_monitor_df = geo_df[geo_df['Site ID'] == nearest_monitor['Site ID']].copy()

# Convert the 'Date' column to a datetime object using pd.to_datetime()
nearest_monitor_df['Date'] = pd.to_datetime(nearest_monitor_df['Date'])

# Extract the month from the 'Date' column
nearest_monitor_df['Month'] = nearest_monitor_df['Date'].dt.month

# Group the filtered DataFrame by month
monthly_stats = nearest_monitor_df.groupby('Month')['Daily Max 1-hour NO2 Concentration'].agg(['mean', 'min', 'max'])

# Reset index
monthly_stats.reset_index(inplace=True)

# Rename columns
monthly_stats.columns = ['Month', 'Monthly Average', 'Monthly Minimum', 'Monthly Maximum']

# Display the monthly statistics
print(monthly_stats)
```

```python
import matplotlib.pyplot as plt

# Create a line plot with markers for the Monthly Average, Minimum, and Maximum NO2 values
plt.plot(monthly_stats['Month'], monthly_stats['Monthly Average'], marker='o', label='Monthly Average')
plt.plot(monthly_stats['Month'], monthly_stats['Monthly Minimum'], marker='o', label='Monthly Minimum')
plt.plot(monthly_stats['Month'], monthly_stats['Monthly Maximum'], marker='o', label='Monthly Maximum')

# Add labels, title, and legend
plt.xlabel('Month')
plt.ylabel('NO2 Concentration')
plt.title('Monthly NO2 Statistics')
plt.legend()

# Set the x-axis ticks to display month numbers
plt.xticks(range(1, 13))

# Display the plot
plt.show()

```
